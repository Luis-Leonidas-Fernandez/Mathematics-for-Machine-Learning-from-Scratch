# Escalares en TensorFlow (versión 2.0 o posterior)

En **TensorFlow 2.x**, un **escalar** es un **tensor de rango 0**, es decir, un tensor sin dimensiones (`shape = []`).

Los escalares (y tensores en general) pueden crearse de distintas formas, tal como se describe en la documentación oficial de TensorFlow:  
https://www.tensorflow.org/guide/tensor

## Creación de escalares

Entre las formas más utilizadas se encuentran:

- `tf.constant` – crea tensores inmutables.
- `tf.Variable` – crea tensores mutables, comúnmente usados como parámetros entrenables.
- `tf.convert_to_tensor` – convierte valores de Python o arrays NumPy en tensores.
- `tf.sparse.SparseTensor` – representa tensores dispersos, generalmente usados en estructuras de mayor dimensión.

En la práctica, **`tf.Variable`** es ampliamente utilizado en contextos de entrenamiento y optimización, ya que permite que el valor del tensor se actualice durante el aprendizaje.

## Comparación con otros frameworks

Al igual que con TensorFlow, en **PyTorch** es posible realizar operaciones equivalentes con tensores y convertirlos fácilmente hacia y desde **arrays NumPy**.

## Tipos de datos

Además, una lista completa de los tipos de datos soportados por los tensores se encuentra disponible en la documentación oficial de TensorFlow:  
https://www.tensorflow.org/api_docs/python/tf/dtypes/DType
